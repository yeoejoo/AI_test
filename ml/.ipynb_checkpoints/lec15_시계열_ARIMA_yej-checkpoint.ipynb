### 1) 정의 
* 과거 데이터의 패턴을 분석하여 미래의 값을 예측하는 방법 
* 과거의 패턴이 미래에도 지속된다는 데이터의 안정성이 기본적인 가정으로 필요

### 2) 시계열 패턴

* <font color=red> $T$ </font> 추세(trend) : 데이터가 장기적으로 증가하거나 감소하는 패턴
* <font color=red> $S$ </font>계절성(seasonality) : 주, 월, 분기, 반기 단위 등 특정 시간의 주기로 나타나는 패턴
* <font color=red> $C$ </font>주기성(cycle) : 경기변동과 같이 정치, 경제, 사회적 요인에 의한 변화로,일정 주기가 없으며 장기적인 변화 현상
* <font color=red> $R$ </font> 불규칙요소(Random, Residual-잔차=이벤트-) : 추세, 순환, 계절성은 뺀 나머지, 돌발적인 요인에 의하여 일어나는 변화로, 예측 불가능한 임이의 변동

### 3) 시계열 패턴 분해
* <b>덧셈 분해(additive decomposition)</b>&nbsp; Trend와 Seasonal이 별개(추세↗ 갑자기 한 기점에 뚝 떨굴 때), $ y_t = S_t+T_t+R_t $
* <b>곱셈 분해(multiplicative decomposition)</b>&nbsp; Trend에 따라 Seasonal이 변화(추세↗ 분산이 점점 커지면서 추세에 비례),  $ y_t = S_t*T_t*R_t $ , 데이터에 0이 존재해서는 안된다.

<font color='blue'>
Additive : 시간에 지남에 따라(Trend가 변화함에 따라) 변동폭이 일정&nbsp; &nbsp; &nbsp; &nbsp;
Multiplicative : Trend가 상승함에 따라 변동폭 증가 </font>

### 4) 정상성 (Stationarity)
* 시간에 따른 평균&분산이 일정한 패턴o -> 그래야 예측 가능 (정상성 화 = 분해한다)


### <font color=red> 5) 비정상성 데이터(랜덤)를 정상성(평균,분산 일정)으로 만드는 방법 2개
*  <font color=red>변동폭이 일정하지 않은 경우 : 로그 변환</font>
*  <font color=red>추세, 계절성이 존재하는 경우 : 차분(y_t - y_{t-1}) </font>

<pre>
* (a)와 같이 시간에 따라 변동폭이 일정하지 않고, 추세와 계절적 영향이 존재하는 비정상 시계열 데이터
* (b)로그변환 --> 변동폭 일정 = 분산 (빨간선 일정)
* (c)차분(가격차/) --> 평균 일정 : a차이가 갈수록 커지니까 -> 차분도 뒤로 갈수록 커질것
* (d)로그변환+차분 --> 평균과 분산이 일정한 정상 시계열로 변환 : 동시에 바꾸면 이렇게 됨(횡보 = 정상성 화)

# 시계열모델

### AR(Autoregression) 자기상관 모델
* <font color=red> 이전의 자신의 관측값을 이용하여 미래의 값을 예측하는 모델 --> 자기 자신의 과거를 사용(자기상관성) ex. 주식가격 회귀식 : (정규분포)+바이어스 +화이트노이즈 </font>
* $Y_t = (Y_{t-1} * w) + …  +c + (e_t * u)$
*  e(t) : white-noise(백색 잡음)  : 일반적인 정규분포(평균이 0, 분산이 1인 정규분포)에서 도출된 random noise 값 (기껏해야 범위 내에서 발생)
*  (X(t-1) * w) + b을 평균으로 가지며 u만큼을 분산으로 가지는 정규분포를 따른다
* <img src="https://www.machinelearningplus.com/wp-content/uploads/2019/02/Equation-1-min.png" width=400>

### MA(Moving Average) 이동평균 모델
* <font color=red>이전 관측값의 예측 오차(e(t-1))를 이용하여 미래 값을 예측하는 모델</font>
* 데이터의 평균값이 시간에 따라 변화하는 추이 
* $Y_t = (e_{t-1} * w) + …  +c + (e_t)$
* <img src="https://www.machinelearningplus.com/wp-content/uploads/2019/02/Equation-2-min.png" width=400>

### ARMA(Autoregressive Moving Average) 모델
* AR + MA 결합
* <img src='https://t1.daumcdn.net/cfile/tistory/99EAB33A5C39D1AD27'>

### ARIMA(p, d, q) 모수 설정
* <font color=red>d차 차분한 데이터에 AR과 MA 모형을 합친 모델 --> AR모형의 Lag : p,  MA모형의 Lag : q, 차분(Diffrence)횟수 : d (max=2 <--권고) </font>

* AR(p): 현재값 <-- p개의 과거값에 의해 예측되는 자기회귀모형
* MA(q) : 현재값 <-- q개의 이전 오차들에 의해 예측되는 이동평균모형
* ARMA(p,q): 현재값 <-- AR(p)와 MA(q)를 함께 사용. 과거 값들과 오차 항들이 동시에 현재 값을 예측
* ARIMA(p, d, q)는 ARMA 모델에 차분(Differencing) 개념을 추가. 차분은 시계열 데이터에서 추세나 계절성 등의 변동 요인을 제거하기 위해 사용. ( p는 자기회귀 차수, d는 차분(Differencing), q는 이동평균 차수 )
  
---------

* AR(p) = ARIMA(p, 0, 0)
* MA(q) = ARIMA(0, 0, q)
* ARMA(p, q) = ARIMA(p, 0, q)

<h3>ACF(Autocorrelation function) :자기상관함수 </h3>
<pre>
* 시차에 따른 관측치들 사이의 관련성(자기상관성)을 측정하는 함수
* 정상계열: 빠르게 0에 수렴, 비정상계열:천천히 감소
</pre>

<h3>PACF(Partial autocorrelation function) :편자기상관함수 </h3>
<pre>
* k 이외의 모든 다른 시점 관측치의 영향력을 배제하고  y_t와 y_t_k 두 관측치의 관련성을 측정하는 함수 </pre>

---
   
* <b>AR</b>의 특성을 띄는 경우, ACF는 천천히 감소하고 <b>PACF</b>는 시차 p이후 급격히 감소 <-- <b>자기상관특징</b>이 쎈 경우, 가까운 3일치 것만 보면 됨
* <b>MA</b>의 특성을 띄는 경우, <b>ACF</b>는 시차 q이후 급격히 감소하고 PACF는 천천히 감소 <-- 큰 추이를 봐야할 경우, 어제 오늘 오차=1,2, 부분이 잘 안보이지만, 장기추세로 보일 때 
* 급격히 감소하는 시차를 각 AR과 MA 모형의 모수(p, q)로 사용

https://www.statsmodels.org/dev/generated/statsmodels.tsa.arima.model.ARIMA.html

## ADF 정상성 검정 : 시계열 데이터가 정상성을 가지는지 여부 확인 
* <b>Augmented Dickey-Fuller test</b>
    * 검정 조건 ( p-value : 5% 이내면 reject으로 대체가설-내말- 선택됨 )
    * <font color=red><b>귀무가설(H0): non-stationary. -"시계열데이터는 비정상성(랜덤워크)을 띈다"-
    * <font color=red><b>대체가설 (H1): stationary. -"아니? 분석가능한 정상성 띄는데?"-

<pre>
statsmodels.tsa.stattools.adfuller(x, maxlag=None, regression='c', autolag='AIC', store=False, regresults=False)

return

The test statistic.

<b>pvalue

* AIC : 손실 가중치 계산 (낮을수록 좋음 / X피쳐를 이용해 Y를 예측할 수 있는 정도 )

### ADF 검정 결과-adfuller-에는 ADF 통계량(ADF Statistic)과 p-value가 포함

* p-value: 이 값은 귀무가설(null hypothesis)에 대한 확률 <--  0.05보다 작으면 (귀무가설을 기각 = '정상성이 있다' )
*                  '정상성이 없다'

# 데이터 정상성 변환
* 분산 : 로그변환
* 평균 : 차분 (23.88 - 25.55) -> (25.55 - 24.44) ->... => shift()
* 차분한 만큼 널값 결측처리 해야 함.
* 널값 : 전후 가격(종가)같이 비슷한 가격으로 채움. = bfill
* 첫날부터 널값이 뜨지 않으려면 최소 7월1일 계산 원하는 날(n)보다 최소 전날의 데이터(n+1)가 있어야 온전히 채워서 알 수 있음 : 더미 데이터 

````python
temp = df[:5]
temp
````

````python
temp.shift(1) # 데이터가 (1)만큼 아래로 하나씩 밀림 / (-1)만큼 거꾸로 한칸 올림
````

````python
#    ↑    :   이후값으로 이전값 채우기
# temp = temp.shift(1)
# # temp.fillna(method='bfill', inplace=True)

#    ↓    : 이전값으로 이후값 채우기 
temp = temp.shift(-1)
temp.fillna(method='ffill', inplace=True)

temp
````

````python
# 차분
````

````python
df_diff = df - df.shift()
df_diff.fillna(method='bfill', inplace=True)   # 아래 데이터를 위로 끌어오기 / ffill : 위 데이터를 아래로 끌어오기
df_diff.head()
````

````python
#차분(1차 차분 : 평균을 일정하게 맞춰놓고)한 데이터(df_diff)로 다시 검정 (adfuller)
````

````python
from statsmodels.tsa.stattools import adfuller
res = adfuller(df_diff,  autolag='AIC')
print( f"{res[1]:.30f}")   # 0.05보다 작다 : 귀무가설기각 대립가설채택(정상성 데이터이다)
````

````python
# 그래프꼬리보면 분산이 좀 날뛰는 듯? 로그변환 해도 되지만, 정상성을 띄고 있기 때문에 더이상 로그변환 하지 않아도 됨.
````

````python
from statsmodels.tsa.seasonal import seasonal_decompose    # 시계열 분해 for 패턴 파악
res = seasonal_decompose(df_diff, model='additive', period=365)
plt.figure(figsize=(12,4))
res.plot()
plt.show()
````

```python
statsmodels.graphics.tsaplots.plot_acf(x, ax=None, lags=None, *, alpha=0.05, use_vlines=True, adjusted=False, fft=False, missing='none', title='Autocorrelation', zero=True, auto_ylims=False, bartlett_confint=True, vlines_kwargs=None, **kwargs)

<pre>
    **kwargs (기타등등인데 중요)
typ (str {'linear', 'levels'}) –
‘linear’ : Linear prediction in terms of the differenced endogenous variables.(예측했을 때, 차분한 것의 데이터-diff-로 받고 싶으면) <--우리는 차분했으니까 linear씀
’levels’ : Predict the levels of the original endogenous variables. (예측했을 때, 답안을 원본데이터-original-로 받고 싶으면)
</pre>

*  <b>linear</b> : 원래 시계열 데이터의 변화량에 대한 예측이 됩니다. 이 방식은 주로 데이터가 선형적인 경향 (<b>주식 가격의 변화량이나 변동률</b>)
*  <b>levels</b> : 차분되지 않은 원래의 값에 대한 예측. <b>주식 가격과 같이 값 자체</b>가 중요한 경우 'levels'를 사용 

````python
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

plt.figure(figsize=(10,3))
ax1 = plt.subplot(1,2,1)
plot_acf(df_diff,   ax=ax1)

ax2 = plt.subplot(1,2,2)
plot_pacf(df_diff,   ax=ax2)

plt.show()
````

<pre> 
# acf : 차분했을 때 -> 그래프(0은 보지 않고, 1부터)가 급.격.히. 정상성 범위 내로 들어오는지 확인 =

# AR의 특성을 띄는 경우, ACF는 천천히 감소하고 PACF는 시차 p이후 급격히 감소 <-- 자기상관특징이 쎈 경우, 가까운 3일치 것만 보면 됨
# MA의 특성을 띄는 경우, ACF는 시차 q이후 급격히 감소하고 PACF는 천천히 감소 <-- 큰 추이를 봐야할 경우, 어제 오늘 오차=1,2, 부분이 잘 안보이지만, 장기추세로 보일 때

AR(p) = ARIMA(p, 0, 0) 
MA(q) = ARIMA(0, 0, q)
ARMA(p, q) = ARIMA(p, 0, q)  
# ARIMA(p, d, q) : p는 자기회귀 차수, d는 차분(Differencing), q는 이동평균 차수   ----AIC가 가장 작은 값을 찾으면 -> p,d,q 최적의 파라미터 알수 있음.

# ARIMA.
* ref: https://www.statsmodels.org/dev/generated/statsmodels.tsa.arima.model.ARIMA.html

```python
class statsmodels.tsa.arima.model.ARIMA(endog, exog=None, order=(0, 0, 0), seasonal_order=(0, 0, 0, 0), trend=None, enforce_stationarity=True, enforce_invertibility=True, concentrate_scale=False, trend_offset=1, dates=None, freq=None, missing='none', validate_specification=True)

````python
from statsmodels.tsa.arima.model import ARIMA         # ARIMA는 AutoRegressive Integrated Moving Average의 약자로, 시계열 데이터 분석과 예측에 널리 사용되는 통계 모델    
from sklearn.model_selection import train_test_split  # 이 함수는 주어진 데이터를 학습 데이터와 테스트 데이터로 분할하는 역할

X_train, X_test = train_test_split(df, test_size=0.01, shuffle=False)  # 전체 데이터(df)를 학습 데이터(X_train)와 테스트 데이터(X_test)로 분할합니다. 이 때 테스트 데이터의 비율은 전체의 1%(test_size=0.01)이며, shuffle=False 옵션은 시계열의 순서를 유지하기 위해 설정
print(X_train.shape, X_test.shape) # 학습 및 테스트 세트의 크기(행과 열 수)를 출력
X_train.index = pd.DatetimeIndex(X_train.index).to_period('D') # X_train의 인덱스를 일별 기간으로 변환합니다('D'는 'Day' 의미). 이렇게 하면 각 관측치가 어느 날짜에 해당하는지 명확하게 알 수 있습니다. 
model=ARIMA(X_train, order=(0, 1, 1)) # ARIMA 모델을 초기화하고 설정값(order=(0, 1, 1))으로 지정하여 X_train에 적합
                                      # order 파라미터에서 (p,d,q) : p: 자기회귀 부분(AR term) / d: 차분 부분(I term) / q: 이동 평균 부분(MA term)


fit_res=model.fit() # 위에서 정의한 ARIMA 모델을 실제로 학습데이터에 적합시키고 그 결과를 fit_res 변수에 저장
fit_res.summary() # 적합된 모델 결과 요약 정보(summary information)를 출력
````
