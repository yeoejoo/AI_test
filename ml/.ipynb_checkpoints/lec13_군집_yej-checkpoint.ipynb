* 군집 : 유사 데이터를 그룹화
* 전반적인 데이터의 유형(특징) 파악에 용이

<pre>
<font size=4><b> 중심점 업데이트(3번)를 위해 평균값을 사용하기 때문에 K-means </b>   
1. 초기점(k) 설정

k는 중심점(centroid)이자, 묶일 그룹(cluster)의 수와 같다.
위 예시에서는 k=3으로 설정(동그라미)
2. 그룹(cluster) 부여

k개의 중심점(동그라미)과 개별 데이터(네모)간의 거리를 측정한다.
가장 가까운 중심점으로 데이터를 부여한다.
3. 중심점(centroid) 업데이트

할당된 데이터들의 평균값(mean)으로 새로운 중심점(centroid)을 업데이트한다.
4. 최적화

2,3번 작업을 반복적으로 수행한다.
변화가 없으면 작업을 중단한다.
결국 아래와 같은 목적함수를 최소화하는 것을 목표로 하는 알고리즘인 것이다.

https://www.linkedin.com/pulse/understanding-k-means-clustering-its-real-use-cases-security-arora

https://blog.kakaocdn.net/dn/eiVxaB/btq5j79KUJh/tK99PexS73gUmbkOJ96Ahk/img.gif

```python
KMeans(n_cluster=, init=, random_state=)
```

* n_cluster : k의 수 / 군집화할 개수, 즉 군집 중심점의 개수를 의미합니다. 사용자가 직접 지정해야 하는 파라미터* 
init : 초기값 지정(default는 random)(배열의 수는 k와 동일해야함 )* 
random_state : random 시드   / 데이터 포인트에서 무작위로 k개를 선택하여 초기 중심점으로 설정설* 정
K-means는 초기값에 따라 군집이 다르게 형성되므로 매번 다른 결과가 나오는 것을 방지하기 위해, 직접 초기값을 지정하거나(init) 시드를 고정(random_state)하는 방법이 있다.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


# ------------------------------------------------------ 훈련데이터
from sklearn.datasets import load_iris
# --------------------------------------------------------------------------
from sklearn.model_selection import train_test_split
# -------------------------------------------------------------------------- 정규화
from sklearn.preprocessing import StandardScaler, MinMaxScaler, RobustScaler
# -------------------------------------------------------------------------- 인코딩
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
# -------------------------------------------------------------------------- 모델
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
# -------------------------------------------------------------------------- 평가
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score 
from sklearn.metrics import confusion_matrix,classification_report
from sklearn.metrics import roc_curve, roc_auc_score
from sklearn.cluster import KMeans
# -------------------------------------------------------------------------- 교차검증

# --------------------------------------------------------------------------

import warnings
warnings.filterwarnings(action="ignore")
````

````python
data = load_iris()
data.keys()
````

````python
df = pd.DataFrame(data=data["data"], columns=data["feature_names"] )  #index=[]
df.columns = ["sl","sw","pl","pw"]
df['target'] = data["target"]
df.head()
````

````python
y = df['target']
X = df.drop('target', axis=1)
print(X.shape, y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y , test_size=0.2, stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape
````

````python
plt.figure(figsize=(7,3))
plt.subplot(1,2,1)    
sns.scatterplot(x="sl", y="sw", data=df, hue="target")
plt.subplot(1,2,2)    
sns.scatterplot(x="pl", y="pw", data=df, hue="target")
sns.scatterplot(x="pl", y="sl", data=df, hue="target")


plt.tight_layout()
plt.show()
````

         0은 따로 구분, 1과2는 섞여있음              /        구분 잘되어 있음(군집)
