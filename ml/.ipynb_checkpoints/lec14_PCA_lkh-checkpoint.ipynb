```python
# ----------------- 학습 ----------------------
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression

# ----------------- 평가 ----------------------
from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.metrics import precision_recall_curve

```

# Data Load

```python
from sklearn.datasets import load_iris
data = load_iris()
data.keys()
df = pd.DataFrame(data=data["data"], columns=data["feature_names"] )  #index=[]
df.columns = ["sl","sw","pl","pw"]
df['target'] = data["target"]
df.head()
```

# 학습 & 평가

## 1차 점수(원본)

```python
y = df['target']
X = df.drop(['target'],  axis=1)
print(X.shape, y.shape)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)

pred  = model.predict(X_test)
print(confusion_matrix(y_test, pred))
accuracy = accuracy_score(y_test, pred)
f1 = f1_score(y_test, pred, average='macro')

print(f"accuracy : {accuracy:.5f}   f1 : {f1:.5f}")
print( classification_report(y_test, pred) ) 
```

```python
from sklearn.decomposition import PCA
from numpy.linalg import eig
```

```python
y = df['target']
X = df.drop(['target'],  axis=1)
print(X.shape, y.shape)
```

# 특이값분해(singular value decomposition;svd)
* explained_variance_score : eigin value 비율의 총합
* def _fit(self, X): https://github.com/scikit-learn/scikit-learn/blob/55bf5d9/sklearn/decomposition/pca.py#L444 
* pca().compont =  -1 * svd()
    * https://stackoverflow.com/questions/44765682/in-sklearn-decomposition-pca-why-are-components-negative


## sklearn explained_variance_ratio_

```python
from numpy.linalg import svd
from sklearn.utils.extmath import svd_flip
mean_ = np.mean(X, axis=0)
print(mean_)
X -= mean_
print(X)

U, S, V = svd(X, full_matrices=False)
# flip eigenvectors' sign to enforce deterministic output
U, V = svd_flip(U, V)
print(U.shape, S.shape, V.shape)

components_ = V

# Get variance explained by singular values
n_samples = X.shape[0]
explained_variance_ = (S ** 2) / (n_samples - 1)
explained_variance_ratio_ = explained_variance_ /  explained_variance_.sum()

print(explained_variance_ratio_)
```

## numpy eig()

```python
X.T
```

```python
W, V = eig(np.cov(X.T))
print(W)
print(V)
```

```python
eig_value = []
for w in W:
    eig_value.append( w / sum(W))
    
print(f"eig_value : {eig_value[:3]}")
print(f"eig_value : {sum(eig_value[:2])}")
```

## PCA explained_variance_ratio_()

```python
pca = PCA(n_components=2, random_state=11) #모든 피처를 특징있는 2개(주성분)로 데이터 묶어줘
X_pca = pca.fit_transform(X)

print(X_pca[:5])
print(pca.explained_variance_ratio_)
print(pca.explained_variance_ratio_.sum())
```

```python
pca_df = pd.DataFrame(X_pca, columns=['pc1', 'pc2'])
pca_df['target'] = y
pca_df.head()
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="pc1", y="pc2", data=pca_df, hue="target")
plt.show()
```

```python
y = pca_df['target']
X = pca_df.drop(['target'], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)

pred  = model.predict(X_test)
print(confusion_matrix(y_test, pred))
accuracy = accuracy_score(y_test, pred)
f1 = f1_score(y_test, pred, average='macro')

print(f"accuracy : {accuracy:.5f}   f1 : {f1:.5f}")
print( classification_report(y_test, pred) ) 
```
