# 학습 & 평가

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression


from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score, precision_score, recall_score
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

## train test 분리

```python
y = df['target']
X = df.drop(['target'],  axis=1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

* stratify=y

```python
y_train.value_counts(), y_test.value_counts()
```

```python
model = RandomForestClassifier(random_state=0)
model.fit(X_train, y_train)
pred  = model.predict(X_test)
proba = model.predict_proba(X_test)
print(pred[:5])
print(proba[:5])

accuracy = accuracy_score(y_test, pred)
print(f"accuracy : {accuracy:.5f}")
```

```python
arr = np.array([0. ,  0.98 , 0.02])
print(arr)
print( np.argmax(arr) )
```

```python
proba_df = pd.DataFrame(proba)
proba_df['pred'] = pred
proba_df.head(10)
```

## 판별함수

```python
model = LogisticRegression(random_state=0)


model.fit(X_train, y_train)
pred  = model.predict(X_test)
proba = model.predict_proba(X_test)
df_var = model.decision_function(X_test)

print(pred[:5])
print(proba[:5])
print(df_var[:5])

dfdf = pd.DataFrame({"pred":pred}) 
dfdf[['proba0','proba1','proba2']] = proba
dfdf[['df0','df1','df2']] = df_var
dfdf.head()

accuracy = accuracy_score(y_test, pred)
print(f"accuracy : {accuracy:.5f}")
```

# 평가 매트릭스

* <font color=red><b> 이진분류 : average='binary'  (디폴트)
* <font color=red><b> 멀티분류 : average= [None, 'micro', 'macro', 'weighted']

* roc_auc_score()
    * average = {'macro'(디폴트) 'micro', 'samples', 'weighted', None
    * average = None  : multi_class='ovr' and average='micro'

* 마이크로 평균(Micro-Average) vs. 매크로 평균(Macro-Averatge)  <br><br>
${macro 평균=\frac{50+60+80+85}{4} = 68.75}$  <br><br>
${micro 평균=\frac{50*50+5*60+2*80+3*85}{60} = 53.58}$

* One-versus-the-rest(OvR) vs. One-versus-One(OvO)
    * One-versus-the-rest(OvR) : 멀티분류
    * One-versus-One(OvO) : 이진분류
  

<img src="https://docs.microsoft.com/en-us/dynamics365/finance/finance-insights/media/payment-prediction-matrix-2.png">

<img src="https://miro.medium.com/v2/resize:fit:720/format:webp/1*VSchph99Wiv6tQpNIvMJbw.png">

<img src="https://images.velog.io/images/cleansky/post/5fd08ca6-7fa7-44fd-9186-48c911e6f759/image.png">

* <font color=red>${ 정확도(accuracy) =  \frac{TN+TP}{TN+FP+FN+TP} } $  : 전체 중 모델이 맞춘 비율 (많이 맞추자)</font>
---    
* <font color=red>${ 재현률(recall) =  \frac{TP}{FN+TP} } $  : 실제P 예측P </font><br><br>
    <font color=red>민감도(${Sensitivity}$)   
    ${TPR(True Positive Ratio)}$</font>  
    암환자분류기 : FN 중요 (<font color=red>소수의 P양성을 잘 맞추자</font>)
---    
* ${ 정밀도(precision) =  \frac{TP}{FP+TP} } $  : 예측P 실제P <br><br>
    스팸메일분류기 : FP 중요
---
* ${ F1 =  \frac{2}{\frac{1}{precision}+\frac{1}{recall}} }  = 2 * \frac{precision*recall}{precision+recall}$  : 정밀도 재현률 조화   


---
* Y축 :   
        ${TPR}$(True Positive Ratio)  
        재현률(${recall}$)  
        민감도(${Sensitivity}$)
---        
* X축 :   
        ${FPR}$(False Positive Ratio)   ${\frac{FP}{TN+FP} } $<br>
        1 - 특이도(${Specificity}$) <br>
        1 - ${TNR}$<br><br><br>
특이도(${Specificity}$)  ${TNR}$(True Negative Ratio) : ${\frac{TN}{TN+FP} } $

---
* ${ROC}$(Receiver Operating Characteristic Curve) : 양성 오답률(맞춘것 틀린것 에 대한 비율)<br>
    X축 변화에 따른 Y축 변화
