# 학습 & 평가

## train test 분리

```python
y = df['target']
X = df.drop(['target'],  axis=1)
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, 
                                                    stratify=y, random_state=1111)
X_train.shape, X_test.shape, y_train.shape, y_test.shape 
```

```python
model = LogisticRegression(random_state=0)
# model = DecisionTreeClassifier(random_state=0)
model.fit(X_train, y_train)

pred  = model.predict(X_test)
proba = model.predict_proba(X_test)
df_var = model.decision_function(X_test)

print(f"y_test : {y_test[:5]}")
print(f"pred : {pred[:5]}")
print(f"proba : {proba[:5]}")
print(f"dec.func : {df_var[:5]}")
# print(np.argmax(np.array([9.99999931e-01, 6.85359405e-08])) )
    
    
accuracy = accuracy_score(y_test, pred)
print(f"accuracy : {accuracy:.5f}")
```

## 임계치(th) 조정
Binarizer : 임계치(threshold=0.5)를 잡고 그 이하이면 0 , 보다크면 1을 반환
<pre>
class sklearn.preprocessing.Binarizer(*, threshold=0.0, copy=True)

```python
from sklearn.preprocessing import Binarizer
```

```python
arr = np.array([0.7, 0.3, 0.5])
brz = Binarizer(threshold=0.5)
print(  brz.fit_transform(arr.reshape(1, -1))  )

brz = Binarizer(threshold=0.2)
print(  brz.fit_transform(arr.reshape(1, -1))  )

brz = Binarizer(threshold=0.8)
print(  brz.fit_transform(arr.reshape(1, -1))  )
```

```python
print(proba_c1[:5])

brz = Binarizer(threshold=0.5)
print(  brz.fit_transform(proba_c1.reshape(-1, 1)) [:5] )

brz = Binarizer(threshold=0.8)
print(  brz.fit_transform(proba_c1.reshape(-1, 1)) [:5] )
```

```python
th_list = [0.35, 0.38, 0.42, 0.5]
for th in th_list:
    print(th)
    brz = Binarizer(threshold=th)
    th_pred = brz.fit_transform(proba_c1.reshape(-1, 1))  
    print(confusion_matrix(y_test, th_pred))
    print(classification_report(y_test, th_pred))
    print("=="*30)
```

## precision_recall_curv vs. roc_curv

* 좌측그림(임계치가 왼쪽(작아진다)으로 간다) == 우측그림(FPR, TPR이 최대) 
    * 임계치가 최소화 == 양성(P) 확률이 커진다
    * 임계치가 0    == 모든 데이터가 양성(P) == TPR(100)
<br><br>

* 좌측그림(임계치가 오른쪽(커진다)으로 간다) == 우측그림(FPR, TPR이 최소) 
    * 임계치가 최대화 == 음성(P) 확률이 커진다
    * 임계치가 1    == 모든 데이터가 음성(N) == FPR(100)

* 왼쪽 그림 : 두 집단간 분류가 잘 된다  == 오른쪽그림 (roc 커브의 면적이 최대화, auc 점수 최대화)
